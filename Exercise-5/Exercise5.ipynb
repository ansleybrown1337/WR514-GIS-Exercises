```python
# set graph size
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams.update({'font.size': 22})
```

```python
# import modules
import geopandas
from geopandas import read_file
import libpysal
from libpysal.examples import explain
import mapclassify
import seaborn as sns
```

```python
# import data
_ = libpysal.examples.load_example('South')
pth = libpysal.examples.get_path('south.shp')
```

```python
df = read_file(pth)
```

## Display histogram of homicide data

```python
explain('South')
```

```python
ax = sns.histplot(data=df, x="HC60", kde=True)
plt.tight_layout()
plt.savefig('homicides_hist.jpg')
```

```python
df.HC60.describe()
```

## Quantiles

```python
ax = df.plot(column='HR60', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},
             edgecolor='k')
ax.set(title='Quantiles: US Southern County Homicides from 1960 to 1990')
plt.tight_layout()
plt.savefig('quantiles.jpg')
```

## Equal Intervals

```python
ax = df.plot(column='HR60', scheme='equal_interval', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},
             edgecolor='k')
ax.set(title='Equal Interval: US Southern County Homicides from 1960 to 1990')
plt.tight_layout()
plt.savefig('equal-interval.jpg')
```

## Natural Breaks

```python
ax = df.plot(column='HR60', scheme='natural_breaks', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)},
             edgecolor='k')
ax.set(title='Natural Breaks: US Southern County Homicides from 1960 to 1990')
plt.tight_layout()
plt.savefig('nat-breaks.jpg')
```
